```python
import torch

!pip install git+https://github.com/BussemakerLab/PyProBound.git
import pyprobound
import pyprobound.plotting
```

```python
alphabet = pyprobound.alphabets.DNA()

# Corrected raw GitHub URLs for direct download
CTCF_r3 = 'https://raw.githubusercontent.com/kiyounghan/dataanalysis26/main/CTCF_r3.100000.tsv'
IMR90_GR = 'https://raw.githubusercontent.com/kiyounghan/dataanalysis26/main/IMR90_GR_chip-seq_rep1.100000.tsv.gz'


dataframe= pyprobound.get_dataframe([CTCF_r3])
dataframe.head()

dataframe2 = pyprobound.get_dataframe([IMR90_GR])
dataframe2.head()

```

```python
count_table = pyprobound.CountTable(
    dataframe,
    alphabet,
    # left_flank="ACACTCTTTCCCTACACGACGCTCTTCCGATCTTGACGTC",
    # right_flank="GACGTCAGATCGGAAGAGCTCGTATGCCGTCTTCTGCTTG",
    left_flank_length=5,
    right_flank_length=5,
    max_left_flank_length=12,
    max_right_flank_length=12,
)
```

```python
nonspecific = pyprobound.layers.NonSpecific(alphabet=alphabet, name="NS")
psam_1 = pyprobound.layers.PSAM(
    kernel_size=12,
    alphabet=alphabet,
    max_kernel_size=18,
    # increment_flank=True,
    # shift_footprint_heuristic=True,
    # increment_footprint=True,
    # increment_flank_with_footprint=True,
)
psam_2 = pyprobound.layers.PSAM(
    kernel_size=13, # Changed kernel_size to make it distinct
    alphabet=alphabet,
    max_kernel_size=18,
    # increment_flank=True,
    # shift_footprint_heuristic=True,
    # increment_footprint=True,
    # increment_flank_with_footprint=True,
)
psams = [psam_1, psam_2]
```

```python
modes = [
    pyprobound.Mode.from_nonspecific(nonspecific, count_table, name="Nonspecific"),
    pyprobound.Mode.from_psam(psam_1, count_table, name="PSAM_1"),
    pyprobound.Mode.from_psam(psam_2, count_table, name="PSAM_2"),
]
```

```python
round_0 = pyprobound.rounds.InitialRound()
round_1 = pyprobound.rounds.BoundUnsaturatedRound.from_binding(modes, round_0)
```

```python
experiment = pyprobound.Experiment(
    [round_0, round_1],
    name="CTCF",
    counts_per_round=count_table.counts_per_round,
)
```

```python
model = pyprobound.MultiExperimentLoss([experiment], pseudocount=20)
```

```python
optimizer = pyprobound.Optimizer(
    model,
    [count_table],
    greedy_threshold=2e-4,
    device="cpu",
    checkpoint="CTCF.pt",
    output="CTCF.txt",
)
```

```python
optimizer.train_sequential()
```
